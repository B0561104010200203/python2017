```python
%matplotlib inline
import matplotlib.pyplot as plt
file = open('data.csv','r')

x=[]
y=[]

while 1:
    data = file.readline()
    if not data:
        break
    buf=data.split(',')
    x.append(float(buf[0]))
    y.append(float(buf[1]))

fig = plt.figure()
plt.plot(x,y,'ro')
plt.show()
fig.savefig('points_1.pdf')

u=0
v=0
ux=0
uy=0

for e in x:
    u+=e
for e in y:
    v+=e

u=u/len(x)
v=v/len(y)

for e in x:
    ux+=(e-u)**2
for e in y:
    uy+=(e-v)**2

ux=(ux/len(x))**0.5
uy=(uy/len(y))**0.5

ss=0

for e in range(len(x)):
    ss+=(x[e]-u)*(y[e]-v)

ly=[ ss/len(x)/ux/ux*(e-u)+v for e in x]

fig = plt.figure()
plt.plot(x,y,'ro')
plt.plot(x,ly)
plt.show()
fig.savefig('line_1.pdf')

zx=[]
zy=[]
for e in x:
    zx.append((e-u)/ux)
for e in y:
    zy.append((e-v)/uy)
    
fig = plt.figure()
plt.plot(zx,zy,'ro')
plt.show()
fig.savefig('points_2.pdf')

r=ss/len(x)/ux/uy

yyy=[r*(e) for e in zx]

fig = plt.figure()
plt.plot(zx,zy,'ro')
plt.plot(zx,yyy)
plt.show()
fig.savefig('line_2.pdf')
file.close()
```
