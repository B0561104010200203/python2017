```python
def hanoi(n, A, B, C):
    if n == 1:
        return [(A, C)]
    else:
        return hanoi(n-1, A, C, B) + hanoi(1, A, B, C) + hanoi(n-1, B, A, C)

n = input("Enter the number of disks")
for move in hanoi(int(n), '1', '2', '3'):
    print("%c -> %c" % move)
```

```python
from collections import deque
from itertools import cycle
 
def hanoi(ndisks):

    tower = [deque(range(ndisks)), deque(), deque()]
 
    pattern = ((0,1),(0,2),(1,2)) if ndisks&1 != 1 else ((0,2),(0,1),(1,2))
 
    for fm, to in cycle(pattern):
        if not tower[fm] or tower[to] and tower[to] < tower[fm]:
            fm, to = to, fm
             
        yield tower[fm][0], fm, to
         
        tower[to].appendleft(tower[fm].popleft())
         
        if len(tower[2]) == ndisks:
            break

format_step = "{} -> {}".format
 

for disk, fm, to in hanoi(5):
    print(format_step('123'[fm], '123'[to]))
```
